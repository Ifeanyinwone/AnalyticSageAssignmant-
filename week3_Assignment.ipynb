```python
import requests
import pandas as pd
from dotenv import load_dotenv
import os

# Step 1: Load environment variables
load_dotenv()

API_KEY = os.getenv("COIN_GECKO_API")

# Step 2: Define API URL
url = "https://api.coingecko.com/api/v3/coins/markets"

# Step 3: Define parameters
params = {
    "vs_currency": "usd",
    "order": "market_cap_desc",
    "per_page": 100,
    "page": 1
}

# Step 4: Define headers (Pro API key)
headers = {
    "x-cg-pro-api-key": API_KEY
}

# Step 5: Send request
response = requests.get(url, params=params, headers=headers)

# Step 6: Check response status
if response.status_code == 200:
    
    # Step 7: Convert to JSON
    data = response.json()
    
    # Step 8: Convert to DataFrame
    df = pd.DataFrame(data)
    
    # Step 9: Select important columns
    df = df[[
        "name",
        "symbol",
        "current_price",
        "market_cap",
        "total_volume"
    ]]
    
    # Step 10: Display result
    print("Top 10 Cryptocurrencies:")
    print(df.head(10))

else:
    print(f"Error: {response.status_code}")
```
