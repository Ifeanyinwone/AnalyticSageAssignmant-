```python
score = int(input("Enter your exam score: "))

# check positive, negative or zero
if score > 0:
    sign = "positive"
elif score < 0:
    sign = "negative"
else:
    sign = "zero"

# check range
if score >= 10 and score <= 100:
    range_check = "within the normal score range (10 - 100)"
else:
    range_check = "outside the normal score range"

# check even or odd
if score % 2 == 0:
    parity = "even"
else:
    parity = "odd"

# final summary
print("Your score is", sign + ",", parity + ", and it is", range_check + ".")
```

```python
# 1. greet function
def greet(name):
    print("Hello " + name + ", welcome to Python class.")

# call the function
greet("Emmanuel")


# 2. calculate_bmi function
def calculate_bmi(weight, height=1.75):
    bmi = weight / (height * height)
    print("BMI is:", bmi)
    return bmi

# test cases
calculate_bmi(70, 1.8)   # both arguments
calculate_bmi(65)        # only weight (uses default height)


# 3. total_score function
def total_score(*args):
    total = sum(args)
    average = total / len(args)
    
    print("Total score:", total)
    print("Average score:", average)
    
    return total, average

# test cases
total_score(50, 60, 70)
total_score(80, 90)
total_score(100, 75, 85, 95)
```

```python
import csv

# write to CSV file
with open("products.csv", "w", newline="") as file:
    writer = csv.writer(file)
    
    # write header
    writer.writerow(["Product", "Price", "Quantity"])
    
    # write rows
    writer.writerow(["Book", 10, 5])
    writer.writerow(["Pen", 2, 20])
    writer.writerow(["Bag", 25, 3])
    writer.writerow(["Shoes", 50, 2])


# read from CSV file
with open("products.csv", "r") as file:
    reader = csv.DictReader(file)
    
    for row in reader:
        print(row)
```

```python
import numpy as np
import pandas as pd

# -------- Part A (NumPy) --------
scores = np.array([45, 60, 75, 30, 90, 55])

print("Array:", scores)
print("Shape:", scores.shape)
print("Data type:", scores.dtype)

print("Mean:", np.mean(scores))
print("Min:", np.min(scores))
print("Max:", np.max(scores))
print("Standard Deviation:", np.std(scores))

print("Scores >= 60:", scores[scores >= 60])


# -------- Part B (Pandas) --------
data = {
    "Name": ["John", "Mary", "David", "Sarah", "Paul"],
    "Age": [20, 21, 19, 22, 20],
    "Score": [45, 65, 55, 30, 80],
    "Grade": ["F", "B", "C", "F", "A"]
}

df = pd.DataFrame(data)

print("\nFirst 3 rows:")
print(df.head(3))

print("\nStudents with Score > 50:")
print(df[df["Score"] > 50])

print("\nSorted by Score (Descending):")
print(df.sort_values(by="Score", ascending=False))

df["Passed"] = np.where(df["Score"] >= 50, True, False)

print("\nWith Passed column:")
print(df)

df.to_csv("class_results.csv", index=False)

new_df = pd.read_csv("class_results.csv")

print("\nReloaded DataFrame:")
print(new_df)
```

```python

import pandas as pd
import matplotlib.pyplot as plt



df = pd.read_csv("eth-usd-max_new.csv")

print("Columns:", df.columns)
print("\nFirst 5 rows:")
print(df.head())

print("\nShape:", df.shape)

print("\nStatistics:")
print(df.describe())


# -------- DATA CLEANING --------
df["snapped_at"] = pd.to_datetime(df["snapped_at"])
df = df.sort_values("snapped_at")
df = df.dropna()


# -------- STEP 2: LINE CHART --------
plt.figure()

plt.plot(df["snapped_at"], df["price"])
plt.title("ETH Price Over Time")
plt.xlabel("Date")
plt.ylabel("Price")

plt.xticks(rotation=45)
plt.tight_layout()

plt.savefig("line_chart.png")
plt.show()


# -------- STEP 3: BAR CHART --------
plt.figure()

plt.bar(df["snapped_at"][:20], df["total_volume"][:20])
plt.title("ETH Volume (First 20 Records)")
plt.xlabel("Date")
plt.ylabel("Volume")

plt.xticks(rotation=45)
plt.tight_layout()

plt.savefig("bar_chart.png")
plt.show()


# -------- STEP 4: PIE CHART --------
plt.figure()

plt.pie(df["price"][:5], labels=df["snapped_at"][:5].dt.date, autopct="%1.1f%%")
plt.title("ETH Price Distribution (First 5 Records)")

plt.savefig("pie_chart.png")
plt.show()
```
