```python
from web3 import Web3
import pandas as pd
from dotenv import load_dotenv
import os

load_dotenv ()
INFURA = os.getenv("INFURA_GATEWAY_URL")
print(f" infura import as {INFURA}")
```

## Connecting to the Blockchain

```python
w3 =  Web3(Web3.HTTPProvider(INFURA))
print(f'Is the connection successful?: {w3.is_connected()}')
```

## Checksum Addresses
# Checksum is a mix of upper and lower case. This as result checking error when using.
EIP

```python
def turn_to_checksum(address: str) -> str:
    if not Web3.is_checksum_address(address):
        return Web3.to_checksum_address(address)
    return address


ACCOUNT_ADDRESS_RAW = "0xE1fC7834857Ece72e1CBaaa227Cf038AE9168267"
ACCOUNT_ADDRESS = turn_to_checksum(ACCOUNT_ADDRESS_RAW)
print(f'Is the address valid?: {Web3.is_checksum_address(ACCOUNT_ADDRESS)}')
```

## Reading Basic Transaction Data

```python
def get_transaction_count(address: str) ->int:
    return w3.eth.get_transaction_count(address)

get_transaction_count(ACCOUNT_ADDRESS)
```

## 1. Get balance of ACCOUNT_ADDRESS

```python
def get_balance(address: str) -> float:
    balance_wei = w3.eth.get_balance(address)
    balance_eth = w3.from_wei(balance_wei, 'ether')
    return balance_eth

print(f"Account balance: {get_balance(ACCOUNT_ADDRESS)} ETH")
```

## 2. Filter transactions by ACCOUNT_ADDRESS

```python
def get_transactions_by_address(address: str, start_block: int, end_block: int):
    transactions = []

    address = address.lower()  # normalize

    for block_num in range(start_block, end_block + 1):
        block = w3.eth.get_block(block_num, full_transactions=True)

        for tx in block.transactions:
            from_addr = tx['from'].lower() if tx['from'] else None
            to_addr = tx['to'].lower() if tx['to'] else None

            if from_addr == address or to_addr == address:
                transactions.append(tx)

    return transactions


# Example usage
latest_block = w3.eth.block_number
txs = get_transactions_by_address(ACCOUNT_ADDRESS, latest_block - 10, latest_block)

print(f"Found {len(txs)} transactions related to address")
```

## 3. Get latest block number

```python
def get_latest_block_number() -> int:
    return w3.eth.block_number

print(f"Latest block number: {get_latest_block_number()}")
```

## 4. Get block details by block number

```python
def get_block_details(block_number: int):
    block = w3.eth.get_block(block_number)
    return block

# Example usage
block_number = get_latest_block_number()
block_details = get_block_details(block_number)

print(block_details)
```
